Las regresiones multilineales de un responden a la expresin

\begin{equation}
\hat{SPI}_{n_{i}} = [a_j]_i^T \cdot [mei_j]_i + b_i 
\end{equation}

siendo $[a_j]_i, [mei_j]_i \in \mathbb{R}^{12 \times 1}$ e indicando $n$ el ndice del SPI e $i$ el mes. 

Por ejemplo, la regresin del SPI$_1$ de enero se calcula por medio de la ecuacin

\begin{equation}
\hat{SPI}_{1_{0}} = a_{0_1} \cdot mei_{E/F}(-1) + \cdots + a_{0_{12}} \cdot mei_{D/E}(0) + b_0 
\end{equation}

Las regresiones fueron hechas aplicando la funcin _LinearRegression_, el cual forma parte del paquete sklearn. Los coeficientes, el trmino independiente y el valor $r^2$ fueron almacenados en listas cuyos nombres se corresponden con la estacin de la que se tomaron los datos y concatenado con la extensin _Regr.

La posicin de un dataframe dentro de la lista refiere al ndice del SPI que se busca aproximar, siendo la posicin 0 el dataframe que contiene los valores relacionados con el $\hat{SPI}_1$, y la posicin 11, un dataframe relacionado con $\hat{SPI}_{12}$.

En un dataframe, las primeras 12 columnas representan los coeficientes de la regresin, la penltima representa los valores del trmino independiente y la ltima los valores de $r^2$. Cada fila se corresponde con un mes en particular, siendo la fila 0 la que contiene los valore de la regresin para el SPI del mes de enero y, la fila 11, la correspondiente a la regresin del SPI para el mes de diciembre.
